```python
dataframe = pd.DataFrame(np.random.randn(100,4), columns= ['a', 'b', 'c', 'd'])
```

```python
dataframe
```

```python
dataframe.plot()
```

```python
dataframe.plot(kind='scatter', x='a', y='b')
```

```python
aggregated_df = dataframe.groupby(['a', 'b'])['c'].sum().reset_index()
pivot_df = aggregated_df.pivot(index='a', columns='b', values='c')
pivot_df.iplot(kind='surface')
```
